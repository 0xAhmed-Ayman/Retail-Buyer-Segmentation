#  feature engineering

```python
spending_cols = ['spend_wine', 'spend_meat', 'spend_fruits',
                 'spend_fish', 'spend_sweets']
df['total_spend'] = df[spending_cols].sum(axis=1)
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['total_accepted_campaigns'], y=df['total_spend'])
plt.title("Total Spend vs Accepted Campaigns")
plt.show()
```

```python
df[['total_accepted_campaigns', 'total_spend']].corr()
```

```python
ref_year = 2014
df['age'] = ref_year - df['birth_year']


def parse_date(s):
    for fmt in ('%d-%m-%Y','%d/%m/%Y','%m/%d/%Y','%d-%m-%y','%d/%m/%y'):
        try:
            return pd.to_datetime(s, format=fmt)
        except:
            continue
    # fallback: let pandas try
    try:
        return pd.to_datetime(s, dayfirst=True, errors='coerce')
    except:
        return pd.NaT

df['signup_date_parsed'] = df['signup_date'].apply(parse_date)
print("Unparsed signup_date:", df['signup_date_parsed'].isna().sum())

df['signup_year'] = df['signup_date_parsed'].dt.year.fillna(0).astype(int)
df['signup_month'] = df['signup_date_parsed'].dt.month.fillna(0).astype(int)
df['signup_dayofweek'] = df['signup_date_parsed'].dt.dayofweek.fillna(0).astype(int)

spend_cols = ['spend_wine','spend_fruits','spend_meat','spend_fish','spend_sweets','spend_gold']
df['total_spend'] = df[spend_cols].sum(axis=1)

for c in spend_cols:
    df[f'{c}_pct'] = df[c] / (df['total_spend'] + 1e-9)
```
